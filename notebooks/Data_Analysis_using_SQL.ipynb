#### Lista de partidos

Se mostraran solo las columnas que nos interesan, por lo que en lugar de * usaremos los nombres exactos. Algunas de las celdas tienen el mismo nombre (Country.name,League.namw). Los renombraremos usando AS.

```python
detalles_partidos = pd.read_sql("""SELECT Match.id, 
                                        Country.name AS country_name, 
                                        League.name AS league_name, 
                                        season, 
                                        stage, 
                                        date,
                                        HT.team_long_name AS  home_team,
                                        AT.team_long_name AS away_team,
                                        home_team_goal, 
                                        away_team_goal                                        
                                FROM Match
                                JOIN Country on Country.id = Match.country_id
                                JOIN League on League.id = Match.league_id
                                LEFT JOIN Team AS HT on HT.team_api_id = Match.home_team_api_id
                                LEFT JOIN Team AS AT on AT.team_api_id = Match.away_team_api_id
                                WHERE country_name = 'Spain'
                                ORDER by date
                                LIMIT 10;""", conn)
detalles_partidos
```

#### Análisis básico

Aquí estamos comenzando a ver los datos a un nivel más agregado. En lugar de mirar los datos sin procesar, comenzaremos a agruparlos en diferentes niveles que queremos examinar. Lo basaremos en la consulta anterior, eliminaremos la información del partido y la fecha, y lo analizaremos a nivel de país-liga-temporada.

```python
ligas_por_temporada = pd.read_sql("""SELECT Country.name AS country_name, 
                                        League.name AS league_name, 
                                        season,
                                        count(distinct stage) AS number_of_stages,
                                        count(distinct HT.team_long_name) AS number_of_teams,
                                        avg(home_team_goal) AS avg_home_team_scors, 
                                        avg(away_team_goal) AS avg_away_team_goals, 
                                        avg(home_team_goal-away_team_goal) AS avg_goal_dif, 
                                        avg(home_team_goal+away_team_goal) AS avg_goals, 
                                        sum(home_team_goal+away_team_goal) AS total_goals                                       
                                FROM Match
                                JOIN Country on Country.id = Match.country_id
                                JOIN League on League.id = Match.league_id
                                LEFT JOIN Team AS HT on HT.team_api_id = Match.home_team_api_id
                                LEFT JOIN Team AS AT on AT.team_api_id = Match.away_team_api_id
                                WHERE country_name in ('Spain', 'Germany', 'France', 'Italy', 'England')
                                GROUP BY Country.name, League.name, season
                                HAVING count(distinct stage) > 10
                                ORDER BY Country.name, League.name, season DESC
                                ;""", conn)
ligas_por_temporada
```

```python
df = pd.DataFrame(index=np.sort(ligas_por_temporada['season'].unique()), columns=ligas_por_temporada['country_name'].unique())

df.loc[:,'Germany'] = list(ligas_por_temporada.loc[ligas_por_temporada['country_name']=='Germany','avg_goals'])
df.loc[:,'Spain']   = list(ligas_por_temporada.loc[ligas_por_temporada['country_name']=='Spain','avg_goals'])
df.loc[:,'France']   = list(ligas_por_temporada.loc[ligas_por_temporada['country_name']=='France','avg_goals'])
df.loc[:,'Italy']   = list(ligas_por_temporada.loc[ligas_por_temporada['country_name']=='Italy','avg_goals'])
df.loc[:,'England']   = list(ligas_por_temporada.loc[ligas_por_temporada['country_name']=='England','avg_goals'])

df.plot(figsize=(12,5),title='Promedio de goles por partido a lo largo del tiempo')
```

```python
altura_de_los_jugadores = pd.read_sql("""SELECT CASE
                                        WHEN ROUND(height)<165 then 165
                                        WHEN ROUND(height)>195 then 195
                                        ELSE ROUND(height)
                                        END AS calc_height, 
                                        COUNT(height) AS distribution, 
                                        (avg(PA_Grouped.avg_overall_rating)) AS avg_overall_rating,
                                        (avg(PA_Grouped.avg_potential)) AS avg_potential,
                                        AVG(weight) AS avg_weight 
                            FROM PLAYER
                            LEFT JOIN (SELECT Player_Attributes.player_api_id, 
                                        avg(Player_Attributes.overall_rating) AS avg_overall_rating,
                                        avg(Player_Attributes.potential) AS avg_potential  
                                        FROM Player_Attributes
                                        GROUP BY Player_Attributes.player_api_id) 
                                        AS PA_Grouped ON PLAYER.player_api_id = PA_Grouped.player_api_id
                            GROUP BY calc_height
                            ORDER BY calc_height
                                ;""", conn)
altura_de_los_jugadores
```

```python
plt.figure(figsize=(12,5))
ax = sns.lineplot(data=altura_de_los_jugadores, x='calc_height', y='avg_overall_rating')
ax.set_title('Potencial vs Altura')
```
